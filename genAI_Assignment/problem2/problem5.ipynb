```python
import random
import pandas as pd
```

```python
# Function to generate random symptom values
def generate_symptom_data(num_records=100):
    data = []
    for _ in range(num_records):
        fever = random.uniform(36.0, 40.0)  # Random body temperature between 36°C and 40°C
        cold = random.randint(0, 10)        # Cold intensity on a scale of 0 to 10
        shivering = random.randint(0, 10)   # Shivering intensity on a scale of 0 to 10
        weight_loss = random.uniform(0, 10) # Weight loss in kg
        data.append({'fever': fever, 'cold': cold, 'shivering': shivering, 'weight_loss': weight_loss})
    return data

# Generate 100 random records
data = generate_symptom_data(100)
```

```python
# Convert the data to a pandas DataFrame
df = pd.DataFrame(data)

```

```python
# Function to order data based on a specific parameter
def order_by_parameter(df, parameter):
    if parameter in df.columns:
        return df.sort_values(by=parameter)
    else:
        return "Invalid parameter!"

# Example of ordering by 'fever'
parameter_to_sort = 'fever'  # Change this to sort by 'cold', 'shivering', or 'weight_loss'
ordered_df = order_by_parameter(df, parameter_to_sort)
```

```python
# Save the sorted data to an Excel file
output_file = "/content/drive/MyDrive/GenAI/covid_symptoms_data.xlsx"
ordered_df.to_excel(output_file, index=False)

print(f"Data saved to {output_file}")
```
