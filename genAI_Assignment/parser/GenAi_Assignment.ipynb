1. Create a Program that uses the Parser
2. Read the DataSet and Create a Dictionary and store it in memory for reference
3. Print the Loss i.e. Dictionary colums those were not found
4. Read the DataSet and enhance the Dictionary for the new data found in other Symptoms

```python
import pandas as pd

# Load the dataset
dataset = pd.read_csv('/content/drive/MyDrive/GenAI/dataset.csv')
```

```python
# Define expected columns for the dictionary
expected_symptoms = ['Fever', 'Cough', 'Cold', 'Other Symptoms']

# Initialize an empty dictionary
symptom_dict = {}

# Initialize a list to track missing data (loss)
missing_data = []
```

```python
# Iterate through each row in the dataset
for index, row in dataset.iterrows():
    patient_id = row['Patient_Id']
    patient_data = {}

    # Create a dictionary entry for each symptom
    for symptom in expected_symptoms:
        if pd.isna(row[symptom]):
            # Track if symptom data is missing
            missing_data.append(f"Missing {symptom} for Patient: {patient_id}")
        else:
            # Add symptom to the dictionary
            patient_data[symptom] = row[symptom]

    # Store the patient data in the main dictionary
    symptom_dict[patient_id] = patient_data

# Print the dictionary
print("Symptom Dictionary:", symptom_dict)

# Print the missing data (loss)
print("Missing Data (Loss):", missing_data)
```

```python
# Read dataset again and enhance dictionary if new data is found
for index, row in dataset.iterrows():
    patient_id = row['Patient_Id']

    # If new data is found, update the dictionary
    for symptom in expected_symptoms:
        if not pd.isna(row[symptom]) and symptom not in symptom_dict[patient_id]:
            symptom_dict[patient_id][symptom] = row[symptom]

# Print the enhanced dictionary
print("Enhanced Dictionary:", symptom_dict)
```
